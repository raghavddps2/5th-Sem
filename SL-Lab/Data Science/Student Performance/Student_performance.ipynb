```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Simply to read the data.
df = pd.read_csv('student.csv')
df.head()
```

```python
#To print the shape of the data
print(df.shape)
```

```python
# To describe the data
df.describe()
```

```python
#Check for any missing values in the dataset.
df.isnull().sum()
```

```python
# We will fill the null values for the 3rd column.
df["parental level of education"].fillna("No college",inplace=True)
```

```python
df.isnull().sum()
```

```python
# Let us explore the math score first
p = sns.countplot(x="mathscore",data=df,palette="bright")
_ = plt.setp(p.get_xticklabels(),rotation=90)
```

```python
# Getting as to how many students passed in math exam
df['Math_Pass_Status'] = np.where(df['mathscore'] < 40,'F','P')
df.Math_Pass_Status.value_counts()
```

```python
# Getting the parental level of education.
p = sns.countplot(x="parental level of education",data=df,palette='bright',hue="Math_Pass_Status") 
_ = plt.setp(p.get_xticklabels(),rotation=90)
```

```python
# Let us explore the reading score
p = sns.countplot(x="reading score",data=df,palette="bright")
plt.show()
```

```python
df["Read_Pass_Status"] = np.where(df["reading score"] < 40,'F','P')
df.Read_Pass_Status.value_counts()
```

```python
p = sns.countplot(x="parental level of education",data=df,hue="Read_Pass_Status",palette="bright")
_ = plt.setp(p.get_xticklabels(),rotation=90)
```

```python
# Let us explore the writing score now
p = sns.countplot(x="writing score",data=df,palette="bright")
plt.show()
```

```python
df["Write_Pass_Status"] = np.where(df["writing score"] < 40, 'F', 'P')
df.Write_Pass_Status.value_counts()
```

```python
p = sns.countplot(x="parental level of education",data=df,hue="Write_Pass_Status")
_ = plt.setp(p.get_xticklabels(),rotation=90)
```

```python
df['OverAll_PassStatus'] = df.apply(lambda x : 'F' if x['Math_Pass_Status'] == 'F' or 
                                    x['Read_Pass_Status'] == 'F' or x['Write_Pass_Status'] == 'F' else 'P', axis =1)
df.OverAll_PassStatus.value_counts()
```

```python
p = sns.countplot(x="parental level of education",hue="OverAll_PassStatus",data=df)
_ = plt.setp(p.get_xticklabels(),rotation=90)
```

```python
df["Total_Marks"] = df["mathscore"] + df["reading score"] + df["writing score"]
df["Percentage"] = df["Total_Marks"]/3
```

```python
p = sns.countplot(data=df,x="Percentage",palette="bright")
_ = plt.setp(p.get_xticklabels(),rotation=90)
```

```python
# Function to define the grade.
def getGrade(Percentage,OverAll_PassStatus):
    
    if OverAll_PassStatus == 'F':
        return 'F'
    elif Percentage >=80:
        return 'A'
    elif Percentage >=70:
        return 'B'
    elif Percentage >=60:
        return 'C'
    elif Percentage >=50:
        return 'D'
    elif Percentage >=40:
        return 'E'
    else:
        return 'E'
    
```

```python
# We have to apply the above function on all the tuples.
df['Grade'] = df.apply(lambda x: getGrade(x["Percentage"],x["OverAll_PassStatus"]),axis=1) #We have to tell we are dealing with rows.
df.Grade.value_counts()
```

```python
p = sns.countplot(x="Grade",data=df,palette="bright",order=['A','B','C','D','E','F'])
_ = plt.setp(p.get_xticklabels(),rotation=90)
```

```python
# Final grade according to the parental level of education.
p = sns.countplot(x="parental level of education",data=df,palette="bright",hue="Grade")
_ = plt.setp(p.get_xticklabels(),rotation=90)
```

```python
# We have the pd.cut, we use it when we wanna segment data into bins. cut will take categores as labels and bins as marks_intervals
marks_intervals = [0,40,50,60,75]
categories = ['failed', 'second class', 'first class', 'distinction']
df["Marks_Categories_Math"] = pd.cut(df.mathscore,marks_intervals,labels=categories)
sns.countplot(x="Marks_Categories_Math",data=df,hue="gender")
```
